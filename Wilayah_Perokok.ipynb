```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Membaca file CSV
data = pd.read_csv('sample_data/Persentase Perokok RI.csv')
```

```python
# Menampilkan 5 baris pertama data untuk memastikan pembacaan file berhasil
print(data.head())
```

```python
# Membuat visualisasi data
plt.figure(figsize=(10, 8))

# Menambahkan bar untuk tahun 2015
plt.barh(data['Provinsi'], data['2015'], color='b', alpha=0.5, label='2015')

# Menambahkan bar untuk tahun 2016
plt.barh(data['Provinsi'], data['2016'], color='r', alpha=0.5, label='2016')

# Menambahkan label sumbu x dan y serta judul
plt.ylabel('Provinsi')
plt.xlabel('Presentasi Perokok')
plt.title('Presentasi Perokok di Indonesia Tahun 2015 dan 2016')
plt.legend()  # Menampilkan legenda

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

```python
# Mengambil top 10 presentasi perokok tertinggi untuk tahun 2015
top_10_2015 = data.nlargest(10, '2015')

# Mengambil top 10 presentasi perokok tertinggi untuk tahun 2016
top_10_2016 = data.nlargest(10, '2016')

# Menggabungkan data top 10 tahun 2015 dan 2016
top_10 = pd.concat([top_10_2015, top_10_2016])

# Membuat visualisasi data
plt.figure(figsize=(10, 6))

# Menambahkan bar untuk tahun 2015
plt.barh(top_10['Provinsi'], top_10['2015'], color='b', alpha=0.5, label='2015')

# Menambahkan bar untuk tahun 2016
plt.barh(top_10['Provinsi'], top_10['2016'], color='r', alpha=0.5, label='2016')

# Menambahkan label sumbu x dan y serta judul
plt.ylabel('Provinsi')
plt.xlabel('Presentasi Perokok')
plt.title('Top 10 Presentasi Perokok di Indonesia Tahun 2015 dan 2016')
plt.legend()  # Menampilkan legenda

# Menampilkan grafik
plt.tight_layout()
plt.show()
```
