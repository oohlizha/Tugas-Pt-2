3. Buatlah list Top 5 Jenis produk yang memiliki nilai transaksi paling banyak menggunakan dataframe pada python!

```python
#Soal 3
# Mengelompokkan data berdasarkan jenis produk dan menjumlahkan nilai transaksi
produk_transaksi = dfmerge.groupby('ProductID')['Amount'].sum()

# Mengurutkan nilai transaksi secara menurun dan mengambil 5 jenis produk teratas
top_5_produk = produk_transaksi.nlargest(5)

# Menampilkan hasil
print("Top 5 Jenis produk dengan nilai transaksi paling banyak:")
print(top_5_produk)
```

4. Buatlah deskripsi statistik untuk masing masing hari (23,24,25,26) pada dataset sample

```python
#Soal 4
dfmerge['Date'] = pd.to_datetime(dfmerge['Date'])

# Filter data untuk masing-masing hari (23, 24, 25, 26)
for hari in [23, 24, 25, 26]:
    data_hari_ini = dfmerge[dfmerge['Date'].dt.day == hari]
    deskripsi_statistik = data_hari_ini.describe()
    print(f"\nDeskripsi Statistik untuk Hari {hari}:")
    print(deskripsi_statistik)
```

5. Carilah waktu terbaik (hari dan jam) dimana paling banyak user gas station melakukan transaksi !


```python
#Soal 5
# Mengubah kolom 'Date' menjadi tipe data datetime
dfmerge['Date'] = pd.to_datetime(dfmerge['Date'])

# Menambahkan kolom 'DayOfWeek' untuk hari dalam seminggu (0: Senin, 1: Selasa, dst.)
dfmerge['DayOfWeek'] = dfmerge['Date'].dt.dayofweek

# Mengelompokkan data berdasarkan hari dalam seminggu dan jam dalam sehari
grouped = dfmerge.groupby(['DayOfWeek', 'Time']).size().reset_index(name='TransactionCount')

# Mencari waktu terbaik berdasarkan jumlah transaksi terbanyak
best_time = grouped.loc[grouped['TransactionCount'].idxmax()]

# Mencetak waktu terbaik
print("Waktu terbaik untuk transaksi:")
print("Hari:", best_time['DayOfWeek'])  # Menggunakan angka 0-6 untuk Senin-Minggu
print("Jam:", best_time['Time'])
print("Jumlah Transaksi:", best_time['TransactionCount'])
```
