```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'sample_data/E-commerce-cosmetic-dataset.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Menampilkan beberapa baris untuk memverifikasi konversi
df.head()
```

```python
# Mengambil 10 produk dengan rating tertinggi
top_10_best_rated = df.nlargest(10, 'rating')[['brand', 'rating', 'price']]

top_10_best_rated
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur ukuran figure
plt.figure(figsize=(12, 10))

# Membuat bar plot untuk 10 produk dengan rating terbaik sesuai brand
sns.barplot(x='rating', y='brand', hue='price', data=top_10_best_rated, dodge=False, palette='viridis')

# Menambahkan judul dan label
plt.title('Top 10 Brand dengan Rating Terbaik dan Harganya')
plt.xlabel('Rating')
plt.ylabel('Brand')

# Menampilkan plot
plt.show()
```
